```python
#Get all relations
hypernyms= set()
hyponyms= set()
count = 0
all_synsets = set()
for w in words:
  w_synsets=iwn.synsets(w,pos=pyiwn.PosTag.NOUN)
  for s in w_synsets:
    all_synsets.add(s)

    #Get all hypernyms
    h1=iwn.synset_relation(s, pyiwn.SynsetRelations.HYPERNYMY)
    for hs1 in h1:
      hypernyms.add((s, hs1))

    #Get all hyponyms
    h2=iwn.synset_relation(s, pyiwn.SynsetRelations.HYPONYMY)
    for hs2 in h2:
      hyponyms.add((s, hs2))
    
    count += 1
    print("\r"+str(count), end= "")
```

```python
print("Total unique synsets: ", len(all_synsets))
print("No of hypernym realtions: ", len(hypernyms))
print("No of hyponym realtions: ", len(hyponyms))
```

```python
#Id to synset mapping
id_2_synset = {}
for hyp in [hypernyms, hyponyms]:
  for h in hyp:
    s1_id = h[0]._synset_id
    s2_id = h[1]._synset_id
    id_2_synset[s1_id] = h[0]
    id_2_synset[s2_id] = h[1]
```

```python
import networkx as nx
G = nx.Graph()
#Directed Graph
# G = nx.DiGraph()
```

```python
#Add all hyperny relation to graph
for h in hypernyms:
  s1_id = h[0]._synset_id
  s2_id = h[1]._synset_id
  s1_head_word = devtrans.dev2iast(h[0]._head_word)  #devtrans coverts the words from Devnagari to roman format
  s2_head_word = devtrans.dev2iast(h[1]._head_word)

  # G.add_node((s1_id, h[0]))
  # G.add_node((s2_id, h[1]))
  G.add_node(s1_id, head_word = s1_head_word)
  G.add_node(s2_id, head_word = s2_head_word)
  #s2 is hypernym of s1 So, s2 --> s1 edge
  G.add_edge(s2_id, s1_id)
```

```python
#Add all hyponym relations to the graph
for h in hyponyms:
  s1_id = h[0]._synset_id
  s2_id = h[1]._synset_id
  s1_head_word = devtrans.dev2iast(h[0]._head_word)
  s2_head_word = devtrans.dev2iast(h[1]._head_word)

  # G.add_node((s1_id, h[0]))
  # G.add_node((s2_id, h[1]))
  # G.add_node(s1_id)
  # G.add_node(s2_id)
  G.add_node(s1_id, head_word = s1_head_word)
  G.add_node(s2_id, head_word = s2_head_word)

  #s2 is hyponym of s1 So, s1-->s2 edge
  G.add_edge(s1_id, s2_id)
```
